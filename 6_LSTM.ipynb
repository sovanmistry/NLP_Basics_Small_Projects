## **Load the IMDB Dataset**

```python
from keras.datasets import imdb
from keras.preprocessing.text import Tokenizer
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data()
```

```python
# Checked the length of different sentences

len(X_train[1]), len(X_train[5])
```

**Conclusion**: *Length is different on every cases hence we need to do Padding*

## **Padding**

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
X_train = pad_sequences(X_train, padding="post", maxlen=50)
X_test = pad_sequences(X_test, padding="post", maxlen=50)
```

**Note :**Maxlen can be Hyperparameter

```python
# Checked the length of different sentences

len(X_train[1]), len(X_train[5])
```

## **Prepare Model**

```python
from keras import Sequential
from keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(10000,output_dim=2,input_length=50))
model.add(LSTM(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
history = model.fit(X_train, y_train, epochs = 5, validation_data= (X_test, y_test))
```

## **Prediction**

```python
test_data = X_test[0][0:50].reshape(1,-1)
```

```python
model.predict(test_data)
```
