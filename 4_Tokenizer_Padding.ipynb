```python
import numpy as np
from keras.preprocessing.text import Tokenizer
```

## **Load Data**

```python
docs = ['recurrent neural network',
		'neural network',
		'artificial neural',
		'connections between nodes',
		'can create a cycle',
		'allowing output',
		'some nodes to affect subsequent',
		'exhibit temporal',
		'dynamic behavior',
		'type of Neural Network',
    'affect subsequent']
```

## **Tokenizer**

```python
tokenizer = Tokenizer(oov_token = '<nothing>')
tokenizer.fit_on_texts(docs)

# the words which are not present in vocabulary will replace with <nothing?
```

## **Index of each and Every Word**

```python
tokenizer.word_index
```

## **Word Count present in document**

```python
tokenizer.word_counts
```

## **Transforms each text in texts to a sequence of integers**

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

**Conclusion : Length of Sequences of each word is not same**

```python
len(sequences[0])
```

```python
len(sequences[1])
```

## **Make All Sequences to the Same Length**

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding = 'post')
```
